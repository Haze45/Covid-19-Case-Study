Analyze the peak number of daily new cases in Germany, France, and Italy. Which country experienced the highest single-day surge, and when did it occur?


```python
countries = ["Germany", "France", "Italy"]
confirmed = confirmed[confirmed["Country/Region"].isin(countries)]
confirmed_grouped = confirmed.groupby("Country/Region").sum(numeric_only=True)
confirmed_grouped = confirmed_grouped.drop(columns=["Lat", "Long"])
daily_new_cases = confirmed_grouped.diff(axis=1)

peak_cases = []
peak_dates = []
peak_info = {}
for country in countries:
    max_cases = daily_new_cases.loc[country].max()
    peak_date = daily_new_cases.loc[country].idxmax()
    peak_cases.append(max_cases)
    peak_dates.append(peak_date)
    peak_info[country] = {"peak_cases": int(max_cases), "date": peak_date}
print(peak_info)
max_country = max(peak_info, key=lambda x: peak_info[x]["peak_cases"])
max_details = peak_info[max_country]
for country, info in peak_info.items():
    print(f"{country}: {info['peak_cases']} cases on {info['date']}")

```

```python
plt.figure(figsize=(10, 6))
bars = plt.bar(countries, peak_cases, color=['skyblue', 'salmon', 'lightgreen'])
plt.title("Peak Daily New COVID-19 Cases")
plt.ylabel("Number of Cases")
```

 Compare the recovery rates (recoveries/confirmed cases) between Canada and Australia as of December 31, 2020. Which country showed better management of the pandemic according to this metric?

```python
countries_CA = ["Canada", "Australia"]
target_date = "12/31/20"

recovery_stats = {}
for country in countries_CA:
    total_confirmed = confirmed[confirmed["Country/Region"] == country][target_date].sum()
    total_recovered = recovery[recovery["Country/Region"] == country][target_date].sum()
    recovery_rate = (total_recovered / total_confirmed) * 100 if total_confirmed > 0 else 0
    recovery_stats[country] = {
        "confirmed": int(total_confirmed),
        "recovered": int(total_recovered),
        "recovery_rate": int(recovery_rate)
    }
print(recovery_stats)
better_country = max(recovery_stats, key=lambda x: recovery_stats[x]["recovery_rate"])
print(better_country)
```

What is the distribution of death rates (deaths/confirmed cases) among provinces in Canada? Identify the province with the highest and lowest death rate as of the latest data point.

```python
confirmed_canada = confirmed[confirmed["Country/Region"] == "Canada"]
deaths_canada = deaths[deaths["Country/Region"] == "Canada"]

# Use the latest date available in the dataset
latest_date = confirmed_canada.columns[-1]

province_data = pd.DataFrame({
    "Province": confirmed_canada["Province/State"].fillna("All Provinces"),
    "Confirmed": confirmed_canada[latest_date].values,
    "Deaths": deaths_canada[latest_date].values
})

# Calculate death rate
province_data["Death Rate (%)"] = (province_data["Deaths"] / province_data["Confirmed"]) * 100
province_data["Death Rate (%)"] = province_data["Death Rate (%)"].fillna(0)

# Find provinces with highest and lowest death rates
highest = province_data.loc[province_data["Death Rate (%)"].idxmax()]
lowest = province_data.loc[province_data["Death Rate (%)"].idxmin()]

# Display results
print("Death Rate by Province in Canada (latest data):")
print(province_data.sort_values("Death Rate (%)", ascending=False))

print(f"\nðŸ”º Highest death rate: {highest['Province']} - {highest['Death Rate (%)']:.2f}%")
print(f"ðŸ”» Lowest death rate: {lowest['Province']} - {lowest['Death Rate (%)']:.2f}%")
```

Question 7: Data Merging
Q7.1: How would you merge the transformed datasets of confirmed cases, deaths, and recoveries on the 'Country/Region' and 'Date' columns to create a comprehensive view of the pandemic's impact?

```python
def melt_data(df, value_name):
    id_vars = ['Province/State', 'Country/Region', 'Lat', 'Long']
    df_melted = df.melt(id_vars=id_vars, var_name='Date', value_name=value_name)
    df_melted['Date'] = pd.to_datetime(df_melted['Date'])
    return df_melted

confirmed_long = melt_data(confirmed, 'Confirmed')
deaths_long = melt_data(deaths, 'Deaths')
recovered_long = melt_data(recovery, 'Recovered')

merged_df = confirmed_long.merge(deaths_long, on=['Province/State', 'Country/Region', 'Lat', 'Long', 'Date'])
merged_df = merged_df.merge(recovered_long, on=['Province/State', 'Country/Region', 'Lat', 'Long', 'Date'])

print(merged_df.head())
```
